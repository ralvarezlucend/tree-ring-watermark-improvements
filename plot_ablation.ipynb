```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

# Reproduction

```python
croppings = {
    "cropping": [10, 30, 50, 70, 90][::-1],
    "tpr@1%fpr": [0.92, 0.69, 0.2, 0.04, 0.07], 
    "auc": [0.9984, 0.9427, 0.8693, 0.7518, 0.6324], 
}
jpeg = {
    "jpeg": [10, 30, 50, 70, 90][::-1],
    "tpr@1%fpr": [1, 1, 0.93, 0.76, 0.42], 
    "auc": [1, 1, 0.9984, 0.9934, 0.9853], 
}
rotation = {
    "rotation": [0, 45, 90, 135, 180, 225, 270, 315],
    "tpr@1%fpr": [1, 0.29, 1, 0.62, 1, 0.38, 0.99, 0.28], 
    "auc": [1, 0.9166, 1, 0.9503, 1, 0.934, 0.9998, .9416 ], 
}
```

```python
sns.set_style(style="darkgrid")
sns.color_palette("tab10")
sns.lineplot(x='cropping', y='auc', data=croppings, label='AUC', marker='o')
sns.lineplot(x='cropping', y='tpr@1%fpr', data=croppings, label='TPR@1%FPR', linestyle='dashed', marker='o')
plt.xlabel('Cropping + Scaling %')
plt.ylabel('AUC/TPR@1%FPR')
plt.ylim(-0.05, 1.05)
plt.gca().invert_xaxis()
plt.title("Cropping + Scaling Ablation")
plt.savefig('plots/cropping_reproduction.png', dpi=600)
```

```python
sns.set_style(style="darkgrid")
sns.color_palette("tab10")
sns.lineplot(x='jpeg', y='auc', data=jpeg, label='AUC', marker='o')
sns.lineplot(x='jpeg', y='tpr@1%fpr', data=jpeg, label='TPR@1%FPR', linestyle='dashed', marker='o')
plt.xlabel('JPEG Compression Quality %')
plt.ylabel('AUC/TPR@1%FPR')
plt.ylim(-0.05, 1.05)
plt.gca().invert_xaxis()
plt.title("JPEG Compression Ablation")
plt.savefig('plots/jpeg_reproduction.png')
```

```python
sns.set_style(style="darkgrid")
sns.color_palette("tab10")
sns.lineplot(x='rotation', y='auc', data=rotation, label='AUC', marker='o')
sns.lineplot(x='rotation', y='tpr@1%fpr', data=rotation, label='TPR@1%FPR', linestyle='dashed', marker='o')
plt.ylim(-0.05, 1.05)
plt.xlabel('Rotation Degree')
plt.ylabel('AUC/TPR@1%FPR')
plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(45))
plt.savefig('plots/rotation_reproduction.png')
```
